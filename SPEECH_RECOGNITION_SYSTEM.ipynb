```python
# ==========================================
# SPEECH RECOGNITION SYSTEM (FIXED VERSION)
# Internship Task - CODTECH
# Using Wav2Vec2 Pretrained Model
# ==========================================

# Install required libraries
!pip install -q transformers torchaudio librosa soundfile

import torch
import librosa
from transformers import AutoProcessor, Wav2Vec2ForCTC
from google.colab import files

print("Upload a short WAV audio file (16kHz recommended)...")
uploaded = files.upload()

file_name = list(uploaded.keys())[0]

# Load processor and model (UPDATED METHOD)
print("\nLoading pretrained Wav2Vec2 model...")
processor = AutoProcessor.from_pretrained("facebook/wav2vec2-base-960h")
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-base-960h")

# Load audio
print("\nProcessing audio...")
speech, rate = librosa.load(file_name, sr=16000)

# Prepare input
inputs = processor(speech, sampling_rate=16000, return_tensors="pt", padding=True)

# Perform inference
with torch.no_grad():
    logits = model(**inputs).logits

# Decode output
predicted_ids = torch.argmax(logits, dim=-1)
transcription = processor.batch_decode(predicted_ids)[0]

print("\n==============================")
print("🎙️ TRANSCRIBED TEXT:")
print("==============================\n")
print(transcription)
```

```python
# Install required library
!pip install -q gTTS pydub

from gtts import gTTS
from pydub import AudioSegment

# Sample text for audio
text = "Artificial Intelligence is transforming the world through innovation and automation."

# Generate speech (MP3)
tts = gTTS(text=text, lang='en')
tts.save("sample.mp3")

# Convert MP3 to WAV (16kHz recommended)
audio = AudioSegment.from_mp3("sample.mp3")
audio = audio.set_frame_rate(16000)
audio.export("sample.wav", format="wav")

print("✅ Sample audio file generated: sample.wav")
```
